```python
from rtlsdr import RtlSdr
import pandas as pd
import seaborn as sns
sdr = RtlSdr()  # type: ignore 
# configure device
sdr.sample_rate = 2.048e6  # Hz
sdr.center_freq = 433e6     # Hz
sdr.freq_correction = 60   # PPM
sdr.gain = 'auto'


s = pd.Series(sdr.read_samples(8192))

sdr.close()
s.plot()
```
