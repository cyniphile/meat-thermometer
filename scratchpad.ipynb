```python
from rtlsdr import RtlSdr
import pandas as pd
```

```python
sdr = RtlSdr()
```

```python
# configure device
sdr.sample_rate = 2.048e6  # Hz
sdr.center_freq = 70e6     # Hz
sdr.freq_correction = 60   # PPM
sdr.gain = 'auto'
```

```python

s = pd.Series(sdr.read_samples(512))
```

```python
s.plot()
```
